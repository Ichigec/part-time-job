## Как видим из графиков и таблицы, пик продаж приходится на "черную пятницу" и новый год ежегодно.

```python
import seaborn as sns 
correlation = df[['Weekly_Sales', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']].corr() 
#выбираем данные которые соответствуют условию
plt.figure(figsize=(15,15)) #Задаем размер рисунка, что бы он влез и был читаем
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix') # Задаем цветовую палитру и остальные настройки

plt.title('Correlation between different fearures') # Указываем название
plt.show() #строим матрицу корреляции, есть слабая корреляция лишь с размером магазина - чем он больше, тем больше продажи.
```

```python
df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(5) # 5 магазинов с наибольшими продажами
```

```python
max_magaz=list(df.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False).head(5).reset_index()['Store'])
# Делаем список из отобранных магазинов
```

```python
for i in max_magaz:
    to_plt=df.loc[df['Store']==i] # Выбираем данные по выбранным магазинам
    plt.plot(to_plt['Date'], to_plt['Weekly_Sales']) #Добавляем каждый магазин отдельно, что бы получились разные цвета
plt.show() #Рисуем график по топ 5 магазинам 
```

```python
to_plt=df.loc[df['Type']=='A'].groupby('Dept')['Weekly_Sales'].sum().sort_values(ascending=False).head(10).reset_index()
#Находим топ 10 отделов среди магазинов типа А по суммарным продажам
to_plt['Dept']=to_plt['Dept'].astype('str') # Меняем тип данных у номеров отделов для корректного отображения на графике
fig, ax = plt.subplots() # указываем объекты для рисования

ax.bar(to_plt['Dept'], to_plt['Weekly_Sales']) # рисуем сам график

#Рисуем задний фон

ax.set_facecolor('seashell') # цвет внутри рамки рисунка
fig.set_facecolor('floralwhite') #цвет вокруг рамки рисунка
fig.set_figwidth(12)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure

plt.show()
```
