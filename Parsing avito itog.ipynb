```python
import requests
from bs4 import BeautifulSoup
import time
import csv
import pandas as pd
import re
def parse_page(search, page):
    q = search
    r = requests.get(f'https://www.avito.ru/moskva?p={page}&q={q}')
    # если надо искать по каким то конкретным категориям - можно изменить запрос например на:
    # https://www.avito.ru/moskva/audio_i_video?p={page}&q={q}, https://www.avito.ru/moskva/rezume?p={page}&q={q}
    soup = BeautifulSoup(r.text, 'html.parser')
    trade_list = soup.find_all('div', {'class': 'item_table-wrapper'}) 
    df2=pd.DataFrame(columns=['Цена','Название','url адрес','Расстояние до метро','Метро'])
    df=pd.DataFrame(columns=['Цена','Название','url адрес','Расстояние до метро','Метро'])

    for i in trade_list:
            price =((i.find('span', {'class': 'snippet-price'}).text))

            price=''.join(re.findall(r'\d', price))
        
            title_obj = (i.find('a', {'class': 'snippet-link'}).text)
            url = 'https://www.avito.ru'+i.find('a', {'class': 'snippet-link'})['href']
            try:
                dist=''.join(re.findall(r'\d',(i.find('span', {'class': "item-address-georeferences-item__after"}).text)))
            except AttributeError:
                dist='Не указанно'
            try:
                metro=i.find('span', {'class': "item-address-georeferences-item__content"}).text
            except AttributeError:
                metro='Не указанно'
            
            df2.loc[1]=[price,title_obj,url,dist,metro]
            df=df.append(df2)
            
    return (df)
```

```python
parse_page('вода', 1)
```
